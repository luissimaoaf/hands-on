## Obtaining the dataset

```python
import os
import tarfile
import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets","housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    # create the directory if it does not yet exist
    os.makedirs(housing_path, exist_ok=True)
    
    # download the dataset
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    
    # extracting the tarfile
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

fetch_housing_data()
```

Loading the dataset into Python.

```python
import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    
    return pd.read_csv(csv_path)

housing = load_housing_data()
housing.head()
```

Visualize the distribution of the data to consider train-test split.

```python
%matplotlib inline
import matplotlib.pyplot as plt

housing.hist(bins=50, figsize=(15,10))
plt.show()
```

We observe non-normal distributions and cut-off points in the data.

```python
import numpy as np

# create categorical income variable
housing["income_cat"] = pd.cut(housing["median_income"], 
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])

housing["income_cat"].hist(density=True)
```

We split training and test sets by stratified sampling to respect categorical distribution.

```python
from sklearn.model_selection import StratifiedShuffleSplit

# stratified sampling
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]

# check if it is correctly stratified
print(strat_test_set["income_cat"].value_counts() / len(strat_test_set))

# drop categorical column
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
    
# setting aside the test set for now
housing = strat_train_set.copy()
```
